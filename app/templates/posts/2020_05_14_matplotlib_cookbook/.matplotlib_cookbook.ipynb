```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats

sns.set(style='ticks', context='poster', color_codes=True)
plt.style.use('style.mplstyle')
!cat style.mplstyle
```

```python
xdata = np.arange(100)
ydata = stats.norm.rvs(loc=0, size=100)

plt.plot(xdata, ydata)

plt.xlabel('x-value')
plt.ylabel('y-value')

plt.savefig('line_plot.png', dpi=600)
```

```python
xdata = np.arange(100)
ydata_1 = stats.norm.rvs(loc=0, size=100)
ydata_2 = stats.norm.rvs(loc=3, size=100)

fig = plt.figure()
plt.plot(xdata, ydata_1, label='y1')
plt.plot(xdata, ydata_2, label='y2')

plt.xlabel('x-value')
plt.ylabel('y-value')
plt.legend(frameon=False, bbox_to_anchor=(0, 1, 1, 0.2), loc='lower right', ncol=2, borderaxespad=0)

plt.savefig('line_plot_multiple.png', dpi=600)
```

```python
xdata = np.arange(100)
ydata_1 = stats.norm.rvs(loc=0, size=(4, 100))
ydata_2 = stats.norm.rvs(loc=3, size=(4, 100))

# Inner Element
nrows, ncols = 2, 2
fig, axs = plt.subplots(nrows, ncols, figsize=(5*1.2*ncols, 3*1.2*nrows), sharex=True, sharey=True)

for i in range(nrows*ncols):
    ax = axs[i//ncols, i%ncols]
    ax.plot(xdata, ydata_1[i], label=f"y1")
    ax.plot(xdata, ydata_2[i], label=f"y2")

# Common legend
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, frameon=False, loc='upper right', ncol=2)

plt.tight_layout(pad=0)
plt.subplots_adjust(top=0.90, wspace=0, hspace=0)

plt.savefig("line_plot_subplots.png", dpi=600)
```
