## Estimation Significance

$$
y = y^* + \epsilon
$$

$$
E(y) = y^* + E(\epsilon)\\
y^* = E(y) - E(\epsilon)
$$

```python
mu = 0
sigma = 3
error_rv = stats.norm(mu, sigma)
error = error_rv.rvs(1000)

plt.hist(error, bins=20);
print(np.mean(error), np.std(error))
```

```python
y_true = 42
y = y_true + error

plt.hist(y, bins=20);
print(np.mean(error), np.std(error))
```

```python
y_mean = np.mean(y)
y_std = np.std(y)
y_true_estimate = y_mean - mu
print(
    f"y = {y_true_estimate:.5g} +/- {y_std/np.sqrt(1000):.5g}"
)
```

## Signal Significance

```python
mu = 0
sigma = 10
error_rv = stats.norm(mu, sigma)
error = error_rv.rvs(1000)
y_true = 42
```

```python
plt.figure(figsize=(20, 4))
plt.plot(error)
plt.title("Error")
```

```python
signal = np.zeros(len(error))
signal_index = 137
signal[signal_index] = y_true

plt.figure(figsize=(20, 4))
plt.plot(signal)
plt.ylim(0, 80)
plt.title("True Signal")
```

```python
y = signal + error

plt.figure(figsize=(20, 4))
plt.plot(y)
plt.title("Measurement")
```

```python
plt.figure(figsize=(20, 4))
plt.plot(y - np.mean(error))
plt.title("Measurement")
```

```python
print(
    f"Signal detected with {(y[signal_index] - mu) / sigma:.5g}-sigma significance"
)
```

```python
print(
    f"y = {y[signal_index]:.5g} ",
    f"with {(signal - mu) / sigma:.5g} sigma significance"
)
```
